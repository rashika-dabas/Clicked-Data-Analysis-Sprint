```python
# Import Dependencies 
import sqlite3 as sql
import pandas as pd
```

# Connect SQL Magic to Database

```python
# Load SQL magic into notebook
%load_ext sql
# %reload_ext sql

# Connect SQL magic backend to the dataset 
%sql sqlite:////kaggle/input/basketball/nba.sqlite
```

```python
# Use the %sql magic command to query the database for the game table and transform the response to a Pandas DataFrame
df_games = %sql select * from game
df_games = df_games.DataFrame() # transform magic response to Pandas DataFrame
df_games.head() # display first five rows of resultant DataFrame
```

# Create Connection Object to Database

```python
# Method - 2
conn = sql.connect('/kaggle/input/basketball/nba.sqlite') # create connection object to database
conn
```

```python
df_games = pd.read_sql('select * from game', conn) # use connection object and a SQL query with Pandas
df_games.head() # display first five rows of resultant DataFrame
```

```python
df_games.columns
```

```python
df_team = pd.read_sql('select * from team', conn) # use connection object and a SQL query with Pandas
df_team.head() # display first five rows of resultant DataFrame
```

```python
df_game_team_details = pd.merge(df_games, df_team, left_on='team_id_home', right_on='id') # join game and team tables
df_game_team_details
```

```python
df_game_team_details.columns
```

```python
df_officials = pd.read_sql('select * from officials', conn) # use connection object and a SQL query with Pandas
df_officials.head() # display first five rows of resultant DataFrame
```

```python
df_game_team_officials_details = pd.merge(df_game_team_details, df_officials, on='game_id') # join df_game_team_details and df_officials tables
df_game_team_officials_details.head()
```

```python
df_game_team_officials_details.columns
```

```python
df_inactive_players = pd.read_sql('select * from inactive_players', conn) # use connection object and a SQL query with Pandas
df_inactive_players.head() # display first five rows of resultant DataFrame
```

```python
df_game_team_officials_inactive_players_details = pd.merge(df_game_team_officials_details, df_inactive_players, on='game_id')
df_game_team_officials_inactive_players_details.head()
```

```python
df_game_team_officials_inactive_players_details.columns
```

# Save Dataset

```python
df_game_team_officials_inactive_players_details.to_csv(
    'df_game_team_officials_inactive_players_details.csv',
    index=False
)
```
